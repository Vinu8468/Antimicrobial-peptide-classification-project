```python
import pandas as pd
from sympy import sequence

data = pd.read_csv("../data/processed/final_dataset_v2.csv")
data.head()
```

```python
AMINO_ACIDS = list("ACDEFGHIKLMNPQRSTVWY")
def compute_aac(sequence):
    length = len(sequence)
    freq = {}
    for aa in AMINO_ACIDS:
        freq[aa] = len(sequence) / length
    return freq
# this will count for all the animo acid in the input sequence and we will use that as the feature.. so there will be 20 feature columns
```

```python
def extract_features(sequence):
    features = {}
    features.update(compute_aac(sequence))
    return features
```
